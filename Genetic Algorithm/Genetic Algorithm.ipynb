```
import sys
import copy
import math
import time
import random
```

```
class Node:

    def __init__(self, name, left_child=None, right_child=None):
        self.name = name
        self.left_child = left_child
        self.right_child = right_child

class FuncGraph:

    def __init__(self):
        self.graph = []
        self.err = 0
        for i in range(0, 15):
            node = Node("")
            self.graph.append(node)

    def random_generate(self):
        for i in range(0, self.graph.__len__()):
            rnd = random.random()
            if i < 7:
                if rnd < 0.4:
                    if rnd < 0.04:
                        self.graph[i].name = "+ math.sin(x) -"
                    elif rnd < 0.08:
                        self.graph[i].name = "- math.sin(x) +"
                    elif rnd < 0.12:
                        self.graph[i].name = "* math.sin(x) +"
                    elif rnd < 0.16:
                        self.graph[i].name = "* math.sin(x) /"
                    elif rnd < 0.2:
                        self.graph[i].name = "/ math.sin(x) *"
                    elif rnd < 0.24:
                        self.graph[i].name = "+ math.cos(x) -"
                    elif rnd < 0.28:
                        self.graph[i].name = "- math.cos(x) +"
                    elif rnd < 0.32:
                        self.graph[i].name = "* math.cos(x) +"
                    elif rnd < 0.36:
                        self.graph[i].name = "* math.cos(x) /"
                    elif rnd < 0.4:
                        self.graph[i].name = "/ math.cos(x) *"
                elif rnd < 0.52:
                    self.graph[i].name = "-"
                elif rnd < 0.64:
                    self.graph[i].name = "+"
                elif rnd < 0.76:
                    self.graph[i].name = "/"
                elif rnd < 0.88:
                    self.graph[i].name = "*"
                elif rnd <= 1:
                    self.graph[i].name = "**"
            else:
                if rnd < 0.4:
                    self.graph[i].name = random.uniform(-10, 10)
                else:
                    self.graph[i].name = "x"

    def print_graph(self):
        for i in range(0, self.graph.__len__()):
            print(self.graph[i].name, end=" ")

    def set_children(self):
        for i in range(0, 7):
            self.graph[i].left_child = self.graph[2 * i + 1]
            self.graph[i].right_child = self.graph[2 * i + 2]

    def in_order_traversal(self, root):
        res = []
        if root:
            res = self.in_order_traversal(root.left_child)
            res.append(root.name)
            res = res + self.in_order_traversal(root.right_child)
        return res
```

```
def bubbleSort(pops):
    for i in range(pops.__len__()):
        for j in range(0, pops.__len__() - i - 1):
            if pops[j].err > pops[j + 1].err:
                pops[j], pops[j + 1] = pops[j + 1], pops[j]


def list_to_str(lst):
    func_string = ''
    for i in range(lst.__len__()):
        if i % 8 == 0:
            func_string += '('
        if i % 4 == 0:
            func_string += '('

        func_string += str(lst[i])

        if i % 4 == 2:
            func_string += ')'
        if i % 8 == 6:
            func_string += ')'
    return func_string


def err_computation(coordinations, pop):
    coordinations = list(coordinations)
    func_pop = eval("lambda x:" + list_to_str(pop.in_order_traversal(pop.graph[0])))
    err = 0
    for i in range(0, coordinations.__len__()):
        err += (func_pop(coordinations[i][0]) - coordinations[i][1]) ** 2
    if isinstance(err, complex):
        err = sys.maxsize
    return err


def cross_over(pop1, pop2):
    new_pop = FuncGraph()
    new_pop.set_children()

    for i in range(0, new_pop.graph.__len__()):
        rnd = random.random()
        if rnd <= 0.5:
            new_pop.graph[i].name = pop1.graph[i].name
        else:
            new_pop.graph[i].name = pop2.graph[i].name
    return new_pop


def mutation(pop):
    new_pop = copy.copy(pop)
    for i in range(0, pop.graph.__len__()):
        rnd = random.random()
        if rnd >= 0.6:
            if i < 7:
                if rnd <= 0.68:
                    new_pop.graph[i].name = "-"
                elif rnd <= 0.76:
                    new_pop.graph[i].name = "+"
                elif rnd <= 0.84:
                    new_pop.graph[i].name = "/"
                elif rnd <= 0.92:
                    new_pop.graph[i].name = "*"
                elif rnd <= 1:
                    new_pop.graph[i].name = "**"
    return new_pop
```

```
if __name__ == '__main__':
    first_population = []
    print("Enter Cords Count : ")
    input_cords_count = int(input())
    cords = []
    err_counter = 0
    counter = 1
    start_time = time.time()
    while input_cords_count != 0:
        print("Enter Coordination " , counter , " (In This Format : x y) : ")
        counter += 1
        temp = input().split(" ")
        temp2 = []
        temp2.append(float(temp[0]))
        temp2.append(float(temp[1]))
        cords.append(temp2)
        input_cords_count -= 1
    print("Coordinations : " , cords)

    while first_population.__len__() < 100:
        population = FuncGraph()
        population.set_children()
        population.random_generate()
        try:
            eval("lambda x:" + list_to_str(population.in_order_traversal(population.graph[0])))
            population.err = err_computation(cords, population)
            err_counter += 1
            first_population.append(population)
        except:
            continue
    generations = 0
    while first_population[0].err > 10 and time.time() - start_time < 40:
        generations += 1
        bubbleSort(first_population)
        first_population = first_population[0:50]
        while first_population.__len__() < 100:
            cross_or_mut = random.random()
            if cross_or_mut <= 0.3:
                mut_rnd = random.randint(0, 49)
                new_pop = mutation(first_population[mut_rnd])
                try:
                    eval("lambda x:" + list_to_str(new_pop.in_order_traversal(new_pop.graph[0])))
                    new_pop.err = err_computation(cords, new_pop)
                    err_counter += 1
                    first_population.append(new_pop)
                except:
                    continue
            else:
                cross_rnd1 = random.randint(0, 49)
                cross_rnd2 = random.randint(0, 49)
                new_pop = cross_over(first_population[cross_rnd1], first_population[cross_rnd2])
                try:
                    eval("lambda x:" + list_to_str(new_pop.in_order_traversal(new_pop.graph[0])))
                    new_pop.err = err_computation(cords, new_pop)
                    err_counter += 1
                    first_population.append(new_pop)
                except:
                    continue
```
